####Functions for drawing smooth distributions  
* I could use a kernel density estimator but these take a long time with as much data as I am using, instead I just fit to a histogram and then smooth it out. 
* This seems to do well for situations where the number of samples is very large.

```python
def smooth_dist(vec):
    h = np.histogram(vec.dropna(), bins=1000, normed=True)
    s = pd.rolling_mean(pd.Series(h[0], index=h[1][1:]), 20)
    return s

def draw_dist(vec, split, ax=None, legend=True):
    _, ax = init_ax(ax)
    for l,v in vec.groupby(split):
        smooth_dist(v).plot(label=l, lw=2, ax=ax)
    if legend:
        ax.legend(loc='upper left', frameon=False)
```

## Initialize Logistic Regression Framework

```python
split2 = study[study != 'HIV Short'] == 'HIV Long'
split2.value_counts()
```

```python
binomial = robjects.r.binomial

hiv = pd.Series(1.*split2, name='HIV')
study_ind = study.replace(r'HIV.*', 's2', regex=True)
study_ind.name = 'study'

cc2 = cell_counts.ix[hiv.index]
cc2 = (cc2.T / cc2.sum(1)).T
cc2 = cc2 - cc2.mean()

df_r = process_factors([age, cc2.NK, cc2.CD4T, cc2.CD8T,
                        cc2.Bcell, cc2.Mono, cc2.Gran,
                        study_ind])
df_r = r_cbind(df_r, hiv)

idx = list(df_r.rownames)
vc = hiv.ix[idx].value_counts()
weights = hiv.ix[idx].map(.5 / vc)
controls = study_ind.ix[idx].ix[ti(hiv==False)].dropna()
vc2 = controls.value_counts()
controls = controls.map((1./3) / vc2) * .5
weights = controls.combine_first(weights) * len(idx)
weights_r = convert_to_r_series(weights)

fmla = formula_r('HIV ~ age + NK + CD4T + CD8T + Bcell + ' + 
                 'Mono + Gran')
m1= robjects.r.glm(fmla, df_r, weights=weights_r, family=binomial)


fmla = formula_r('HIV ~ feature + age + study + NK + CD4T + CD8T + ' + 
                 'Bcell + Mono + Gran')

def regression_p(probe, get_model=False):
    feature = logit_adj(combined_beta.ix[probe, study.index])
    feature = (feature - feature.mean()) - feature.std()
    feature.name = 'feature'

    df_f = r_cbind(df_r, feature)
    m = robjects.r.glm(fmla, df_f, weights=weights_r, family=binomial)
    p_val = robjects.r.anova(m1, m, test='F')[5][1]
    if get_model:
        return p_val, m
    else:
        return p_val
    
#%time rr = pd.Series({p: regression_p(p) for p in combined_beta.index})
#mv_model = rr
#mv_model.to_csv('/cellar/users/agross/Desktop/probes_regression_9_12.csv')

#mv_model = pd.read_csv('/cellar/users/agross/Desktop/probes_regression_7_18.csv',
#                       header=None, index_col=0, squeeze=True)
```

```python
residuals = convert_robj(robjects.r.residuals(m1))
```

```python
#Do not import
violin_plot_pandas(study, residuals)
```

```python
%%time
w = weights / sum(weights)
y = residuals - sum(residuals * w)
X = combined_beta.ix[:, w.index].copy()
X = X.subtract(X.multiply(w).sum(1), axis=0)

#Variance
vX = (X * X * w).sum(1)
vY = (y * y * w).sum()

#covariance
vXY = (X * y * w).sum(1)

#correlation
cc = vXY / np.sqrt(vX * vY)

resid_corr = cc
```
