```python
def age_plot(vec, s, flip):
    if flip:
        s = -(s + 1)
    p = vec.index[s]
    sns.set(style="ticks", context="talk")
    df = pd.concat([residual.clip(-15,15), df_norm.ix[p], hiv],
                   keys=['error',p,'HIV'],
                   axis=1)
    pal = {0:"#6495ED", 1:"#F08080"}
    sns.lmplot("error", p, hue="HIV", data=df,
               hue_order=[0,1], palette="muted", size=5,
               legend_out=False)
    plt.gca().set_title(str(probe_annotations.ix[p]))
    

interact(age_plot, vec=widgets.fixed((a2.abs() - a1.abs()).dropna().order()), 
         s=widgets.IntSliderWidget(start=0, min=0, max=50), 
         flip=widgets.CheckboxWidget(values=[True, False]))
```
