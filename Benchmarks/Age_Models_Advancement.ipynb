## Across Study Variation

Look at differential methylation of age-matched samples across the different control cohorts

```python
import igraph as ig

def generate_matching(a1, a2, d=0):
    g = ig.Graph()
    g.add_vertices(list(a1.index))
    g.add_vertices(list(a2.index))
    g.add_edges([(a,b) for a,i in a1.iteritems() for b,j in a2.iteritems()
                    if abs(i - j) <= d])
    matching = g.maximum_bipartite_matching(types=([True] * len(a1)) + ([False] * len(a2)))
    m = pd.Series(matching.matching, g.vs['name'])
    s = pd.Series(range(0,len(g.vs['name'])), g.vs['name'])
    sr = pd.Series(g.vs['name'], range(0,len(g.vs['name'])))
    match = m.map(sr).dropna()
    return match
```
