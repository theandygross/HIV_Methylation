####Functions for drawing smooth distributions  
* I could use a kernel density estimator but these take a long time with as much data as I am using, instead I just fit to a histogram and then smooth it out. 
* This seems to do well for situations where the number of samples is very large.

```python
def smooth_dist(vec):
    h = np.histogram(vec.dropna(), bins=1000, normed=True)
    s = pd.rolling_mean(pd.Series(h[0], index=h[1][1:]), 20)
    return s

def draw_dist(vec, split, ax=None, legend=True):
    _, ax = init_ax(ax)
    for l,v in vec.groupby(split):
        smooth_dist(v).plot(label=l, lw=2, ax=ax)
    if legend:
        ax.legend(loc='upper left', frameon=False)
```

```python
def smooth_dist(vec):
    h = np.histogram(vec.dropna(), bins=300, normed=True)
    s = pd.rolling_mean(pd.Series(h[0], index=h[1][1:]), 20, center=True)
    return s

def draw_dist(vec, split, ax=None, legend=True, colors=None):
    _, ax = init_ax(ax)
    for l,v in vec.groupby(split):
        if colors is None:
            smooth_dist(v).plot(label=l, lw=2, ax=ax)
        else:
            smooth_dist(v).plot(label=l, lw=2, ax=ax, color=colors[l])
    if legend:
        ax.legend(loc='upper left', frameon=False)
```
